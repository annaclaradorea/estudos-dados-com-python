### üìò | Hipotese 01: Criando um DataFrame - Sal√°rio

````python
df_salarios = pd.DataFrame({
'nome': ['Ana','Barbara','Celia','Danilo','Val'],
  'salario': [5000, np.nan, 10000, np.nan,100000]
})

df_salarios.head()
````

### üìÑ | Tratando valores ausentes - utilizando c√°lculos:

````python
#Subtitui os valores ausentes pela M√âDIA dos valores presentes.
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#Subtitui os valores ausentes pela MEDIANA dos valores presentes.
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
````

---
üë£ | PASSO A PASSO:
* *criou uma nova coluna ('salario_media'),*
* *pegou a coluna salario da base (df_salario['salario']),*
* *preencheu os nulos (fillna) com a m√©dia (mean) da coluna salario,*
* *arredonda com duas casas decimais (round(2)).*
---

###üìò | Hipotese 02: Criando um DataFrame - Temperatura

````python
df_temperaturas = pd.DataFrame({
  "dia":['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
  'temperatura': [30,np.nan,np.nan,28,27]
})

df_temperaturas.head()
````

### üìÑ | Tratando valores ausentes - substituindo por valores presentes:

````python
#Substituindo pelo valor anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

#Substituindo pelo valor posterior
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

df_temperaturas
````

üåû | *ffil -> forward fill; bfill -> back fill*

### üìÑ | Criando gr√°ficos interativos (com Plotly)

**üìä | Gr√°fico de barras interativo - M√©dia de sal√°rios por n√≠vel de experi√™ncia**

````python
# Calcula a m√©dia salarial por senioridade
media_por_senioridade = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

# Criar o gr√°fico de barra
fig = px.bar(media_por_senioridade, x='senioridade', y='usd',
             title='M√©dia salarial por senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Sal√°rio m√©dio anual (USD)'}
)


fig.show()
````

O .groupby() deixa a senioridade no √≠ndice da Series, n√£o como uma coluna.

O .reset_index() transforma esse √≠ndice de volta para uma coluna normal do DataFrame. Porque bibliotecas como o Plotly (px.bar) precisam que a coluna que voc√™ quer usar no eixo X/Y esteja como coluna normal, n√£o como √≠ndice.

````python
import pycountry

#Fun√ß√£o para converter ISO-2 para ISO-3:
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

#Criar nova coluna com c√≥digo ISO-3:
df_limpo['empresa_iso3'] = df_limpo['empresa'].apply(iso2_to_iso3)

#Calcular m√©dia sal√°rial por pa√≠s:
salario_por_pais = df_limpo[df_limpo['cargo'] == 'Data Scientist'].groupby('empresa_iso3')['usd'].mean().sort_values(ascending=False).reset_index()

#Gerar o mapa
fig = px.choropleth(
    salario_por_pais,
    locations='empresa_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Sal√°rio m√©dio de Cientistas de Dados por PA√çS',
    labels={'empresa_iso3': 'Pa√≠s',
            'usd': 'Sal√°rio m√©dio anual (USD)'},
)

fig.show()
````
