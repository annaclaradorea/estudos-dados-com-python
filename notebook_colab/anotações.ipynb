### 📘 | Hipotese 01: Criando um DataFrame - Salário

````python
df_salarios = pd.DataFrame({
'nome': ['Ana','Barbara','Celia','Danilo','Val'],
  'salario': [5000, np.nan, 10000, np.nan,100000]
})

df_salarios.head()
````

### 📄 | Tratando valores ausentes - utilizando cálculos:

````python
#Subtitui os valores ausentes pela MÉDIA dos valores presentes.
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#Subtitui os valores ausentes pela MEDIANA dos valores presentes.
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
````

---
👣 | PASSO A PASSO:
* *criou uma nova coluna ('salario_media'),*
* *pegou a coluna salario da base (df_salario['salario']),*
* *preencheu os nulos (fillna) com a média (mean) da coluna salario,*
* *arredonda com duas casas decimais (round(2)).*
---

###📘 | Hipotese 02: Criando um DataFrame - Temperatura

````python
df_temperaturas = pd.DataFrame({
  "dia":['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
  'temperatura': [30,np.nan,np.nan,28,27]
})

df_temperaturas.head()
````

### 📄 | Tratando valores ausentes - substituindo por valores presentes:

````python
#Substituindo pelo valor anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

#Substituindo pelo valor posterior
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

df_temperaturas
````

🌞 | *ffil -> forward fill; bfill -> back fill*

**📊 | Histograma - Distribuição dos salários anuais**

❔ | **O que é histograma?**

---

❕ | Um histograma é um gráfico que mostra a distribuição de frequência dos dados contínuos.Ele divide os dados em intervalos (chamados de "bins") e mostra quantos valores caem em cada intervalo, representando isso com barras verticais.

É útil para visualizar padrões, como se os dados estão concentrados, espalhados, ou se têm alguma forma específica (simétrica, assimétrica, etc).

````python
plt.figure(figsize=[8,5]) #1° número: Largura, 2° número: Altura

#Pedindo um histograma
sns.histplot(df_limpo['usd'], bins = 50, kde=True) #bins: largura da barra, kde: curva que mostra a a tendência geral da distribuição

#Definindo o título geral do gráfico
plt.title('Distribuição dos salários anuais')

#Definindo o rótulo para o eixo X
plt.xlabel('Salário')

#Definindo o rótulo para o eixo Y
plt.ylabel('Frequência')

plt.show()
````

**📊 | Boxplot - Distribuição dos salários anuais (01)**

❔ | **O que é Boxplot?**

---

❕ | Um boxplot é um gráfico que mostra a distribuição de um conjunto de dados, destacando cinco pontos principais: **mínimo, primeiro quartil (Q1), mediana (Q2), terceiro quartil (Q3) e máximo**. Ele ajuda a visualizar rapidamente a mediana, a variação, a simetria dos dados e possíveis valores atípicos (outliers).

É muito usado para comparar distribuições entre grupos diferentes de forma clara e compacta.

````python
plt.figure(figsize=[8,5])

#Pedindo um boxplot
sns.boxplot(x=df_limpo['usd'])

#Definindo o título geral do gráfico
plt.title('Distribuição dos salários anuais')

#Definindo o rótulo para o eixo X
plt.xlabel('Salário em USD')

plt.show()
````

### 📄 | Criando gráficos interativos (com Plotly)

**📊 | Gráfico de barras interativo - Média de salários por nível de experiência**

````python
# Calcula a média salarial por senioridade
media_por_senioridade = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

# Criar o gráfico de barra
fig = px.bar(media_por_senioridade, x='senioridade', y='usd',
             title='Média salarial por senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Salário médio anual (USD)'}
)


fig.show()
````

O .groupby() deixa a senioridade no índice da Series, não como uma coluna.

O .reset_index() transforma esse índice de volta para uma coluna normal do DataFrame. Porque bibliotecas como o Plotly (px.bar) precisam que a coluna que você quer usar no eixo X/Y esteja como coluna normal, não como índice.

**📊 | Gráfico de pizza interativo - Proporção por tipo de trabalho**

````python
contagem_remoto = df_limpo['remoto'].value_counts().reset_index()
contagem_remoto.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(contagem_remoto,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de remoto')
fig.show()
````

**📊 | Gráfico de rosca interativo - Proporção por tipo de trabalho**

````python
contagem_remoto = df_limpo['remoto'].value_counts().reset_index()
contagem_remoto.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(contagem_remoto,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de remoto',
             hole=0.5
             )
fig.update_traces(textinfo= 'percent+label')
fig.show()
````

### 🏹 | DESAFIO: Criar um gráfico de salário médio por país para o cargo de Cientista de Dados

### ➡ | Tentativas:

````python
#Salario medio por pais - Filtrando Data Scientist
salario_por_pais = df_limpo[df_limpo['cargo'] == 'Data Scientist'].groupby('empresa')['usd'].mean().sort_values(ascending=False).reset_index()

salario_por_pais
````

````python
plt.figure(figsize=[12,5])

fig = px.bar(
    salario_por_pais[:15],
    x='empresa',
    y='usd',
    title='Salário médio de Cientistas de Dados por PAÍS (Top 15)',
    labels={'empresa': 'País',
            'usd': 'Salário médio anual (USD)'},
    color='empresa',
    color_discrete_sequence=px.colors.qualitative.Alphabet,
)



fig.show()
````

````python
plt.figure(figsize=[12,5])

#Criar um grafico de salário médio por País em formato de mapa

fig = px.choropleth(
    salario_por_pais,
    locations='empresa',
    color='usd',
    hover_name='empresa',
    title='Salário médio de Cientistas de Dados por PAÍS',
    labels={'empresa': 'País',
            'usd': 'Salário médio anual (USD)'},
)




fig.show()
````

````python
import pycountry

#Função para converter ISO-2 para ISO-3:
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

#Criar nova coluna com código ISO-3:
df_limpo['empresa_iso3'] = df_limpo['empresa'].apply(iso2_to_iso3)

#Calcular média salárial por país:
salario_por_pais = df_limpo[df_limpo['cargo'] == 'Data Scientist'].groupby('empresa_iso3')['usd'].mean().sort_values(ascending=False).reset_index()

#Gerar o mapa
fig = px.choropleth(
    salario_por_pais,
    locations='empresa_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Salário médio de Cientistas de Dados por PAÍS',
    labels={'empresa_iso3': 'País',
            'usd': 'Salário médio anual (USD)'},
)

fig.show()
````
